```python
import csv
import matplotlib.pyplot as plt

drivedir = "/content/drive/Shareddrives/JCC/TCC/"
savedir = "/content/drive/Shareddrives/JCC/TCC/Graficos/"

def plotLoss():
    train_loss = []
    val_loss = []
    val_acc = []

    with open(drivedir + 'Resultados/losses_25.csv' , 'r') as csvFile:
        reader = csv.reader(csvFile)

        for row in reader:
            train_loss.append(float(row[0]))
            val_loss.append(float(row[1]))
            val_acc.append(float(row[2]))

        csvFile.close()

    x = [i for i in range(len(train_loss))]

    plt.figure(figsize = (8, 4))

    plt.title('Perda x Época')
    plt.xlabel('Época')
    plt.ylabel('Perda')

    plt.xticks(x)
    plt.locator_params(axis = 'x', nbins = 10)

    plt.plot(x, train_loss, label = 'Perda de Treinamento', color = 'blue')
    plt.plot(x, val_loss, label = 'Perda de Validação', color = 'red')

    plt.legend()

    plt.savefig(savedir + 'loss.jpg')
    plt.show()

    # -------------------------------------------------------------------
    # FAZER O GRÁFICO COM A ACC DO TESTE

    plt.figure(figsize = (8, 4))

    plt.title('Acurácia x Época')
    plt.xlabel('Época')
    plt.ylabel('Acurácia')

    plt.xticks(x)
    plt.locator_params(axis = 'x', nbins = 10)

    plt.plot(x, val_acc, label = 'Valor da Acurácia', color = 'blue')

    plt.legend()

    plt.savefig(savedir + 'val_acc.jpg')
    plt.show()


plotLoss()
```
